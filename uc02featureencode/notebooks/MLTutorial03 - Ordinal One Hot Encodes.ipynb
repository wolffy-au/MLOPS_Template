```python
# https://machinelearningmastery.com/how-to-prepare-categorical-data-for-deep-learning-in-python/
# 1 of 2
# load and summarize the dataset
from libmlops.data.data_loading import load_csv_data, explore_dataset

...
filename = '../data/external/breast-cancer.csv'
# load the dataset as a pandas DataFrame
names = [
    'age',
    'menopause',
    'tumor-size',
    'inv-nodes',
    'node-caps',
    'deg-malig',
    'breast',
    'breast-quad',
    'irradiat',
    'class',
]
dataset = load_csv_data(filename, names)

# retrieve numpy array
# explore_dataset(dataset)

# shape
# dataset.shape
print(dataset[:3], dataset.shape)
```

```python
...
# prepare input data
from libmlops.features.feature_engineering import (
    ordinal_encode_categorical_features,
    one_hot_encode_categorical_features,
)

# ordinal_encode_categorical_features(dataset, dataset.columns[:-1])
dataset_encoded = one_hot_encode_categorical_features(
    dataset,
    [
        'age',
        'menopause',
        'tumor-size',
        'inv-nodes',
        'node-caps',
        'deg-malig',
        'breast',
        'breast-quad',
        'irradiat',
    ],
)

X_enc = dataset_encoded.drop(columns='class')

print(X_enc.columns, X_enc.shape)
X_enc
```

```python
from libmlops.data.data_preprocessing import split_train_test_xy

# split into input (X) and output (y) variables
# split into train and test sets
X_train, X_test, y_train, y_test = split_train_test_xy(X_enc, y_enc, random_state=1)

# summarize
print("Train", X_train.shape, y_train.shape)
print("Test", X_test.shape, y_test.shape)
print(X_test[:1])
print(y_test[:1])
```

```python
...
# define the model
from sklearn.svm import SVC
from sklearn.ensemble import HistGradientBoostingClassifier
from libmlops.models.model_training import train_model
from libmlops.models.model_evaluation import evaluate_classifier_model

# model = SVC(gamma='auto')
model = HistGradientBoostingClassifier()
train_model(model, X_train, y_train)

# evaluate the keras model
accuracy, report = evaluate_classifier_model(model, X_test, y_test)
print("Accuracy score: ", accuracy, "\n")
print("Classification report:\n", report)
```

```python
from libmlops.models.model_evaluation import (
    confusion_matrix_model,
    plot_confusion_matrix,
)

cm = confusion_matrix_model(model, X_test, y_test)
print("Confusion Matrix:\n", cm)
# plot_confusion_matrix(cm, model)
```
